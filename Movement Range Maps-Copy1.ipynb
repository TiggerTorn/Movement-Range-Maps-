```python
def plot_monthly_avg(df, regions, date_column, value_column):
    plt.figure(figsize=(12, 6))

    for region in regions:
        region_df = df.query(f"polygon_name == '{region}'").copy()
        region_df[date_column] = pd.to_datetime(region_df[date_column])
        region_df.set_index(date_column, inplace=True)
        region_m_avg = region_df[value_column].resample('M').mean()
        region_m_avg.plot(label=region)

    plt.xlabel("Date")
    plt.ylabel(value_column)
    plt.title("Weekly Average by Region")
    plt.legend(loc='lower right')  # Move legend to the right side
    plt.show()
```

```python
def plot_daily_avg(df, regions, date_column, value_column, start_date, end_date):
    plt.figure(figsize=(12, 6))

    # Generate a date range for shading weekends
    all_dates = pd.date_range(start=start_date, end=end_date)
    weekends = all_dates[all_dates.weekday > 4]  # 5 and 6 correspond to Saturday and Sunday

    for region in regions:
        region_df = df.query(f"polygon_name == '{region}'").copy()
        region_df[date_column] = pd.to_datetime(region_df[date_column])
        region_df.set_index(date_column, inplace=True)
        region_df = region_df[start_date:end_date]
        region_d_avg = region_df[value_column].resample('D').mean()
        region_d_avg.plot(label=region)

    # Highlighting weekends
    for weekend in weekends:
        plt.axvspan(weekend, weekend + pd.Timedelta(days=1), color='gray', alpha=0.2)

    plt.xlabel("Date")
    plt.ylabel(value_column)
    plt.title("Daily by Region with Weekends Highlighted")
    plt.legend(loc='lower right')
    plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=7))  # Setting x-axis ticks to show every week
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
THA = THA.reset_index()
THA['ds'] = pd.to_datetime(THA['ds'])
```

```python
THA['day_of_week'] = THA['ds'].dt.dayofweek
```

```python
THA_weekdays = THA[THA['day_of_week'] <= 4]
THA_weekends = THA[THA['day_of_week'] >= 5]
```

```python
# Metrics' average for weekdays
avg_weekdays = THA_weekdays[['all_day_bing_tiles_visited_relative_change', 'all_day_ratio_single_tile_users']].mean()

# Metrics' average for weekends
avg_weekends = THA_weekends[['all_day_bing_tiles_visited_relative_change', 'all_day_ratio_single_tile_users']].mean()

# Combining the data
df_combined = pd.DataFrame({'Weekdays': avg_weekdays, 'Weekends': avg_weekends})

# Plotting
df_combined.plot(kind='bar', figsize=(10,6))
plt.title('Comparison between Weekdays and Weekends')
plt.ylabel('Average Value')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
